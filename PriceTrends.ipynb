```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import os
```

```python
#enter the url
url='https://agmarknet.gov.in/SearchCmmMkt.aspx?Tx_Commodity=20&Tx_State=UC&Tx_District=0&Tx_Market=0&DateFrom=01-May-2025&DateTo=29-May-2025&Fr_Date=01-May-2025&To_Date=29-May-2025&Tx_Trend=0&Tx_CommodityHead=Mango&Tx_StateHead=Uttrakhand&Tx_DistrictHead=--Select--&Tx_MarketHead=--Select--'
page=requests.get(url) #check if you can scrape the data
```

```python
#get the html of the page
soup=BeautifulSoup(page.text,'html')

#find the table headers for columns
table_headers=soup.find_all('th')
table_header=[item.text.strip() for item in table_headers]

#create a dataframe for storing the data
df=pd.DataFrame(columns=table_header)
df
```

```python
#extract all the records from the table
row_tag=soup.find_all('tr')
for text in row_tag:
    table_rows=[item.text.strip() for item in text.find_all('td')]
```

```python
#add these rows to the dataframe
length=len(df)
for data in soup.find_all('tr')[1:]:
    individual_row_data=[item.text.strip() for item in data.find_all('td')]
    length=len(df)
    df.loc[length]=individual_row_data
df
```
